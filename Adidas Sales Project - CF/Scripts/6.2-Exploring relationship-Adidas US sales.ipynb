### 2.1 Correlation Matrix

```python
#Create a correlation matrix using pandas
df.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize = (10,10))

corr = sns.heatmap(df.corr(), annot = True, ax = ax)
```

#### 'Retailer ID' seems to have no significant correlation with any other quantitative variables apart from 'Price per Unit'. It could be worth it to explore this relationship further as maybe some retailers sell products for lower prices or they only sell cheaper product categories.
#### There seems to be a cluster of strong positive correlation between 'Total Sales', 'Units Sold' and 'Operating Profit'.It makes sense as they are related in terms of calculation of revenue and profit.


### 2.2 Scatterplot

```python
# Create a scatterplot for 'Total Sales' and 'Operating Profit' in seaborn
sns.lmplot(x = 'Total Sales', y = 'Operating Profit', data = df)
```

#### The trend line shows strong positive relationship. There is a big concentration at the bottom left corner. However, as we move to the left the the variance of some values increases as well and are further from the trend line.

### 2.4 Categorical plot

```python
# Create a histogram to visualize the distribution of the variable and determine sensible categories for the price ranges
sns.histplot(df['Price per Unit'], bins = 20, kde = True)
```

```python
df.loc[df['Price per Unit'] < 30, 'Price Category'] = 'Low price'
```

```python
df.loc[(df['Price per Unit'] >= 30) & (df['Price per Unit'] < 70), 'Price Category'] = 'Middle price'
```

```python
df.loc[df['Price per Unit'] >= 70, 'Price Category'] = 'High price'
```

```python
df['Price Category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style = "ticks")
g = sns.catplot(x = "Units Sold", y = "Price per Unit", hue = "Price Category", data = df)
```

#### As expected the 'Middle price' category is most pupolated.
#### What is interesting is that there are few outliers that have high number of units sold despite the high price.
